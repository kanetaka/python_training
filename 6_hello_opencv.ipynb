# OpenCVを使って画像を読込、表示、保存する

```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("test.jpg")
```

```python
# 画像の次元数
print("(h, w, c)=", image.shape)
```

```python
# OpenCVのカラーの並びはBGRなので、下記コードだと、、、
plt.imshow(image)
plt.show()
```

## 【問題】元の色のままで表示するプログラムを作成してください

```python
# ここにコードを書く
image_rgb = image[:,:,::-1]
```

```python
plt.imshow(image_rgb)
plt.show()
```

## 【問題】色反転させて、画像を表示させるプログラムを書いてください。

```python
hoge = cv2.bitwise_not(image_rgb)
plt.imshow(hoge)
plt.show
```

## 上記で色反転さ
せた画像を保存するプログラムを書いてください

```python
cv2.imwrite("output.jpg", hoge[:,:,::-1])
```
